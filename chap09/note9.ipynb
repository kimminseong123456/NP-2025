# 1) 컴퓨터 네트워크란?
- 정보와 자원 공유를 위해 링크로 연결된 노드(컴퓨터)들의 집합

## 1-1) 컴퓨터 네트워크 구성 요소 
1. 메시지(message)
    - 메시지는 패킷으로 구성
    - Payload(적화물): 문자, 숫자, 소리, 영상, 그림 또는 멀티미디어 데이터등 전송하는데 필요한 헤더 등을 제외한 순수한 데이터를 의미한다.
2. 송신자(Sender)
    - 메시지를 보내는 장치
3. 수신자(Receiver)
    - 메시지를 받는 장치
